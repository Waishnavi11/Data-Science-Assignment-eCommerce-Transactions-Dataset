```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score, silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Load Datasets
customers = pd.read_csv(r"C:\Users\ajayc\Downloads\Customers.csv")
transactions = pd.read_csv(r"C:\Users\ajayc\Downloads\Transactions.csv")

# Merge data (if applicable)
data = pd.merge(customers, transactions, on='CustomerID')

# Preprocessing: Handle missing values, feature scaling
data.fillna(0, inplace=True)  # Example: fill missing values with 0
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.select_dtypes(include=[np.number]))  # Only numerical columns

# Applying KMeans Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_data)

# Predict clusters
data['Cluster'] = kmeans.labels_

# Evaluate Clustering
db_index = davies_bouldin_score(scaled_data, kmeans.labels_)
sil_score = silhouette_score(scaled_data, kmeans.labels_)

# Print metrics
print(f"DB Index: {db_index}")
print(f"Silhouette Score: {sil_score}")

# Visualize clusters
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=data['Cluster'], palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# Optionally, you can also plot the cluster centers
centroids = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red')
plt.show()
```
